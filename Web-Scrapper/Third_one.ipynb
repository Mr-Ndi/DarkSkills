```python
from bs4 import BeautifulSoup
import requests
```

```python
url='https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_united_State_by_revenue'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
tables = soup.find_all('table')
if len(tables) > 1:
    table = tables[1]
else:
    print(" Not enough tables found on the page.")
```
